Given a random binary generator, write functions to generate an integer sampled from:
1. a uniform distrubition with maximum value 'max_val'.
2. a gaussian distrubition with mean 'mu'.

```python
# import necessary modules
import random
import matplotlib.pyplot as plt
```

```python
# binary random generatory, P(x=1) = p
def random_bit_generator(p):
    '''generates 1 or 0 with p being the probability of 1.'''
    if random.random() < p:
        return 1
    else:
        return 0
```

## Gaussian number:
the sum 1's and 0's produced from a set of N bernouli experiments, or a binomial experiment, has a gaussian disribution, assuming that N is large.

```python
def gaussian_random_number_generator(mu):
    '''generates a random integer number sampled from a gaussian distrubition
    with mean of mean_val.'''
    number = 0
    for i in range(2*mu):\
        number += random_bit_generator(0.5)
    return number
```

## Uniform Distrubition:
we can generate 1's and 0's for an integer number in a binary format and get a uniform distribution.

```python
def uniform_random_number_generator(max_val):
    max_bit_nums = len(bin(max_val))-2
    number = 0
    for i in range(max_bit_nums):
        number += random_bit_generator(0.5)*2**i
    # return number only if smaller or equal to max_val, otherwise
    # run again.
    if number <= max_val:
        return number
    else:
        return uniform_random_number_generator(max_val)
```
